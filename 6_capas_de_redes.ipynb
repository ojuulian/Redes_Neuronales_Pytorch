* import torch.nn as nn  ------>  Para usar el módulo Neural Networks

No obstante se procede a instanciar tres capas:

* nn.Linear(in_features=4096, out_features=10) ------>  Una capa lineal con argumentos in_features (tamaño tensor de entrada), out_features (Tamaño de tensor de salida)

* nn.Conv2d: Capa convolucional 2d para imágenes y 3d para videos:
  - _in_channels:_ Canales de entrada en este caso 3.
  - _out_channels:_ aqui va el numero de canales que se quieren en la salida.
  - _kernel_size:_ Por defecto es 3 sive para hacer los computos.
  - _stride:_ 
  - _padding:_

```
import torch
import torch.nn as nn     #torch.nn:  de neural Networks

linear = nn.Linear(in_features=4096, out_features=10)           # la capa lineal solo toma tensores de dim 1 input
conv = nn.Conv2d(in_channels=3, out_channels=9, kernel_size=3, stride=1, padding=1) #modificar los canales, no va a modificar tamaño de imagen
relu = nn.ReLU(False)
```

```
x = torch.randn(4096)
y = linear(x)
y.size()
```

Una capa lineal lo que hace es multiplicar por una matriz (Transformación lineal). Esta capa Linear tambien se le denomina fully conected

```
x = torch.randn(1,3,7,7)   #1 imagen ,3 canales, (7,7) pixeles de tamaño 
y = conv(x)
z = relu(x)
```
