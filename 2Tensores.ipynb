Por el momento la TPU no estÃ¡ disponible, luego hay que configurar el notebook **(entorno de ejecucion)>>(TPU)**

```
import torch
torch.cuda.is_available()
```

```
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```
x = torch.randn(3,5) #cpu
y = torch.ones_like(x, device=device) #gpu
```

```
x = x.to(device) #gpu
```

```
z = x + y
```

```
print(z)
print(z.cpu())
```
